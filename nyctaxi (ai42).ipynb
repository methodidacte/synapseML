# Get data from Data Lake

```
years_path = [
            "/mnt/delta/nyctaxi/parquet/2019/**",
            "/mnt/delta/nyctaxi/parquet/2020/**",
            "/mnt/delta/nyctaxi/parquet/2021/**",
            "/mnt/delta/nyctaxi/parquet/2022/**"
]
```

```
sdf = spark.read.parquet(*years_path)
```

```
sdf.count()
# 2019 - 2022Q1 : 149 223 088
# 2018 - 2022Q1 : 252 094 475
```

```
display(sdf)
```

```
sdf.printSchema()
```

### clean data

```
from pyspark.sql import functions as F

sdf_clean = sdf.select("VendorID", "trip_distance", "passenger_count", "RatecodeID", "payment_type", "PULocationID", "DOLocationID", "store_and_fwd_flag", "fare_amount") \
    .filter(F.col("trip_distance")>0) \
    .filter(F.col("passenger_count").isNotNull()) \
    .filter(F.col("fare_amount")>0) \
    .filter(F.col("fare_amount")<50000)

display(sdf_clean)
```

```
del sdf
```

```
sdf_clean.count()
# 2019 - 2022Q1 : 144 240 949
```

#SynapseML

```
import synapse.ml
```

```
labelCol = "fare_amount"
```

### train test split

```
train, test = sdf_clean.randomSplit([0.75, 0.25], seed=42)
```

## TrainedRegressorModel

```
from synapse.ml.train import TrainRegressor, TrainedRegressorModel
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import StringIndexer
```

```
help(synapse.ml.train.TrainRegressor)
```

```
lr = LinearRegression().setRegParam(0.1).setElasticNetParam(0.3)
model = TrainRegressor(model=lr, labelCol=labelCol).fit(train)
```

```
model_name = "dbfs:/mnt/delta/nyctaxi/modelMML/nycTaxiModel.mml"
model.write().overwrite().save(model_name)
```

### score and evaluate

```
nycTaxiModel = TrainedRegressorModel.load(model_name)

scoredData = nycTaxiModel.transform(test)
display(scoredData.limit(10).toPandas())
```

```
from synapse.ml.train import ComputeModelStatistics

metrics = ComputeModelStatistics().transform(scoredData)
metrics.toPandas()
```

```
from synapse.ml.train import ComputePerInstanceStatistics

evalPerInstance = ComputePerInstanceStatistics().transform(scoredData)
evalPerInstance.select("fare_amount", "prediction", "L1_loss", "L2_loss").limit(10).toPandas()
```

##LightGMB

```
from pyspark.ml.feature import VectorAssembler
from synapse.ml.lightgbm import LightGBMRegressor
```

```
help(synapse.ml.lightgbm.LightGBMRegressor)
```

### Train

```
feature_cols = sdf_clean.columns[:-1]
featurizer = VectorAssembler(
    inputCols=feature_cols,
    outputCol='features'
)
train = featurizer.transform(train)[labelCol, 'features']
test = featurizer.transform(test)[labelCol, 'features']
```

```
model = LightGBMRegressor(objective='regression',
                          labelCol=labelCol,
                          alpha=0.2,
                          learningRate=0.3,
                          numLeaves=31).fit(train)
```

### Feature importance

```
print(model.getFeatureImportances())
```

```
import pandas as pd
import matplotlib.pyplot as plt

feature_importances = model.getFeatureImportances()
fi = pd.Series(feature_importances,index = feature_cols)
fi = fi.sort_values(ascending = True)
f_index = fi.index
f_values = fi.values
 
# print feature importances 
print ('f_index:',f_index)
print ('f_values:',f_values)

# plot
x_index = list(range(len(fi)))
x_index = [x/len(fi) for x in x_index]
plt.rcParams['figure.figsize'] = (20,5)
plt.barh(x_index,f_values,height = 0.10 ,align="center",color = 'tan',tick_label=f_index)
plt.xlabel('importances')
plt.ylabel('features')
plt.show()
```
